Detecting outliers Using the Interquartile Range (IQR)

```python
# Define a function to detect outliers using IQR
def detect_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]


# Check for outliers in specific columns
for col in ["ModA", "ModB", "WS", "WSgust"]:
    outliers = detect_outliers_iqr(data[col])
    print(f"Outliers in {col}:\n{outliers}")
```

Handling oultiers

```python
def cap_outliers(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column.clip(lower_bound, upper_bound)

# Apply capping
data[["ModA", "ModB", "WS", "WSgust"]] = data[["ModA", "ModB", "WS", "WSgust"]].apply(cap_outliers)
print("Data after capping outliers:")
data.head()
```

Time siries analaysis

```python

# Convert Timestamp to datetime and set as index
data['Timestamp'] = pd.to_datetime(data['Timestamp'])
data.set_index('Timestamp', inplace=True)

# Resample data to observe patterns by hour or day
data_hourly = data.resample('H').mean()
data_daily = data.resample('D').mean()

# Plot GHI, DNI, DHI, and Tamb over time
plt.figure(figsize=(12, 6))
data[['GHI', 'DNI', 'DHI', 'Tamb']].plot(ax=plt.gca(), title="Solar Irradiance and Temperature Over Time")
plt.xlabel('Time')
plt.ylabel('Values')
plt.legend(['GHI', 'DNI', 'DHI', 'Tamb'])
plt.grid()
plt.show()

# Impact of Cleaning on ModA and ModB
cleaned = data[data['Cleaning'] == 1]
not_cleaned = data[data['Cleaning'] == 0]

plt.figure(figsize=(12, 6))
plt.plot(cleaned.index, cleaned['ModA'], 'r-', label='ModA (Cleaned)')
plt.plot(not_cleaned.index, not_cleaned['ModA'], 'b--', label='ModA (Not Cleaned)')
plt.title('Impact of Cleaning on ModA Sensor Readings')
plt.xlabel('Time')
plt.ylabel('Sensor Reading')
plt.legend()
plt.grid()
plt.show()

plt.figure(figsize=(12, 6))
plt.plot(cleaned.index, cleaned['ModB'], 'r-', label='ModB (Cleaned)')
plt.plot(not_cleaned.index, not_cleaned['ModB'], 'b--', label='ModB (Not Cleaned)')
plt.title('Impact of Cleaning on ModB Sensor Readings')
plt.xlabel('Time')
plt.ylabel('Sensor Reading')
plt.legend()
plt.grid()
plt.show()
```

Correlation analysis

```python
# Correlation Matrix
correlation_matrix = data[["GHI", "DNI", "DHI", "TModA", "TModB", "WS", "WSgust", "WD"]].corr()

# Heatmap for Correlation Matrix
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Matrix")
plt.show()

# Pair Plot for Solar Irradiance and Temperature
sns.pairplot(data, vars=["GHI", "DNI", "DHI", "TModA", "TModB"], diag_kind="kde", kind="scatter")
plt.show()

# Scatter Plot for Wind Conditions and Solar Irradiance
plt.figure(figsize=(10, 6))
sns.scatterplot(data=data, x="WS", y="GHI", label="WS vs GHI")
sns.scatterplot(data=data, x="WSgust", y="GHI", label="WSgust vs GHI")
plt.title("Wind Conditions vs Solar Irradiance")
plt.xlabel("Wind Speed")
plt.ylabel("GHI")
plt.legend()
plt.grid()
plt.show()
```

                                            wind analysis

```python
from windrose import WindroseAxes
from matplotlib import cm

# Replace 0 WD with NaN for plotting
data['WD'] = data['WD'].replace(0, np.nan)

# Radial Bar Plot: Wind Speed Distribution by Direction
plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.bar(data['WD'], data['WS'], normed=True, opening=0.8, edgecolor='white', cmap=cm.coolwarm)  # Correct cmap object
ax.set_legend(title="Wind Speed (m/s)")
plt.title("Wind Speed Distribution by Direction")
plt.show()

# Wind Rose: Frequency and Intensity of Wind
plt.figure(figsize=(8, 8))
ax = WindroseAxes.from_ax()
ax.contourf(data['WD'], data['WS'], bins=np.arange(0, 1.1, 0.2), cmap=cm.viridis)  # Correct cmap object
ax.set_legend(title="Wind Speed (m/s)")
plt.title("Wind Rose - Wind Speed and Direction")
plt.show()
```
